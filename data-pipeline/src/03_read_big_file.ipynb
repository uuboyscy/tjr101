```python
"""
select *
from some_table
where portno = '2998623'
and eff_dt = '2021/01/08';
"""
filter_portno_2998623 = df["portno"] == "2998623"
filter_eff_dt_20210108 = df["eff_dt"] == "2021/01/08"

df[
    filter_portno_2998623 & filter_eff_dt_20210108
]
```

```python
"""
select *
from some_table
where portno = '2998623'
and eff_dt = '2021/01/08'
and (percent_tna > 2 or percent_tna < 1);
"""
filter_portno_2998623 = df["portno"] == "2998623"
filter_eff_dt_20210108 = df["eff_dt"] == "2021/01/08"
filter_percent_tna_gt_2 = df["percent_tna"] > 2
filter_percent_tna_lt_1 = df["percent_tna"] < 1

df[
    filter_portno_2998623
    & filter_eff_dt_20210108
    & (
        filter_percent_tna_gt_2 | filter_percent_tna_lt_1
    )
]
```

```python
"""
select *
from some_table
where portno = '2998623'
and eff_dt = '2021/01/08'
and (percent_tna <= 2 or percent_tna >= 1);
"""
filter_portno_2998623 = df["portno"] == "2998623"
filter_eff_dt_20210108 = df["eff_dt"] == "2021/01/08"
filter_percent_tna_gt_2 = df["percent_tna"] > 2
filter_percent_tna_lt_1 = df["percent_tna"] < 1

df[
    filter_portno_2998623
    & filter_eff_dt_20210108
    & ~(
        filter_percent_tna_gt_2 | filter_percent_tna_lt_1
    )
]
```
